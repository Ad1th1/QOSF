### MATH

1. Probability Theory
    > Used to describe the likelihood of different outcomes in a quantum measurement.\
    > The probability distribution of quantum states is used to generate random numbers, ensuring true randomness as opposed to pseudo-randomness.

1. Linear Algebra: 
    > Crucial for describing quantum states and their transformations. \
    > Quantum states are represented as vectors in a Hilbert space, and operations on these states are performed using matrices.\ 
    > For example, quantum gates, which manipulate qubits, are represented by unitary matrices.

1. Complex Numbers:
    > Quantum states are often described using complex numbers, as the amplitudes of quantum states can be complex.\
    > These amplitudes determine the probability distribution of measurement outcomes.\
    > The use of complex numbers allows for the representation of phenomena like interference and phase shifts.
    
1. Statistical Analysis:
    > Statistical methods are used to analyze and verify the randomness of the numbers generated by QRNGs.\
    > Techniques such as randomness tests (e.g., NIST test suite) ensure that the generated numbers meet the criteria for true randomness, crucial for applications in cryptography and security.

### QUANTUM JARGON

- **Qubit**:
  > A quantum bit that can be in a state of 0, 1, or both simultaneously.
- **Superposition**:
  > The ability of a qubit to be in multiple states at once.
- **Entanglement**:
  > A phenomenon where qubits become interconnected, so the state of one instantly affects the state of another.
- **Teleportation**:
  > The process of transferring a qubit's state to another qubit at a different location without moving the physical particle.
- **Quantum Gate**:
  > A basic operation that changes the state of qubits, similar to classical logic gates.
- **Quantum Circuit**:
  > A sequence of quantum gates that perform a specific quantum computation.
- **Measurement**:
  > The process of observing a qubit's state, which collapses it from a superposition to a definite state (0 or 1).
- **Shor's Algorithm**:
  > A quantum algorithm that efficiently factors large numbers, threatening classical encryption methods.
- **Grover's Algorithm**:
  > A quantum algorithm that speeds up the search process in unsorted databases.



#### Applications of Random Number Generators?
> Random number generators are crucial for encryption, simulations, and gaming, ensuring unpredictable and fair outcomes.

#### What We Are Using Today is Pseudo random number generators?
> Currently, most applications use pseudo-random number generators (PRNGs), which rely on deterministic algorithms, potentially making them predictable.

#### Why QRNG Can Be a Game Changer?
> Quantum random number generators (QRNGs) use quantum processes to generate true randomness, enhancing security and reliability in critical applications.

```python
# Cryptomite

import cryptomite

def generate_qrng_numbers(count):
    # Initialize the QRNG
    qrng = cryptomite.QRNG()

    # Generate 'count' quantum random numbers
    random_numbers = [qrng.random() for _ in range(count)]
    
    return random_numbers

def main():
    # Number of quantum random numbers to generate
    num_of_random_numbers = 10

    # Generate the quantum random numbers
    qrng_numbers = generate_qrng_numbers(num_of_random_numbers)

    # Print the generated quantum random numbers
    for index, number in enumerate(qrng_numbers):
        print(f"Random Number {index + 1}: {number}")

if __name__ == "__main__":
    main()
```
